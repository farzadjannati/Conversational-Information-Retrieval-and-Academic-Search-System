سپس مقدار p@1 را برای نتایج جستجو به ازای هر کاربر محاسبه کرده و در انتها این معیار و میانگینش را برای کاربران نمایش دهید.

🎯 <b>خروجی مورد انتظار:</b><br>
جستجوی معنایی برای کوئری‌های اولیه کاربران، محاسبه p@1 و نمایش آن، محاسبه میانگین p@1 برای همه کاربران

````python
# WRITE YOUR CODE HERE
# Perform semantic search for each user's initial query and compute P@1

def semantic_search(query: str, table, embed_model, top_k: int = 5) -> pd.DataFrame:
    """
    Perform semantic search in the vector database using the query embedding.
    Returns top_k most similar documents.
    """
    # Encode the query
    query_embedding = embed_model.encode(
        query, convert_to_numpy=True, normalize_embeddings=True
    ).tolist()
    
    # Search in LanceDB
    results = table.search(query_embedding).limit(top_k).to_pandas()
    return results


def compute_p_at_1(retrieved_paper_id: str, ground_truth_ids: list) -> int:
    """
    Compute Precision@1: 1 if the top retrieved document is in ground truth, else 0.
    """
    return 1 if retrieved_paper_id in ground_truth_ids else 0


# Open the vector database table
db = lancedb.connect(DB_PATH)
table = db.open_table(TABLE_NAME)

# Run semantic search for each user's initial query
print("=" * 80)
print("BASELINE SEMANTIC SEARCH WITH INITIAL QUERIES")
print("=" * 80)

p_at_1_scores = []

for user in users_data:
    user_id = user.get("persona_id", "unknown")
    initial_query = user["initial_query"]
    
    # Extract ground truth relevant paper IDs from the relevant_papers list
    ground_truth_list = [p["paper_id"] for p in user["relevant_papers"]]
    
    # Perform semantic search
    results = semantic_search(initial_query, table, embed_model, top_k=5)
    
    # Get top-1 result
    top_1_paper_id = results.iloc[0]["paper_id"]
    
    # Compute P@1
    p1 = compute_p_at_1(top_1_paper_id, ground_truth_list)
    p_at_1_scores.append(p1)
    
    # Display results for each user
    print(f"\nUser {user_id} ({user['name']}):")
    print(f"  Initial Query: {initial_query}")
    print(f"  Ground Truth Papers: {ground_truth_list}")
    print(f"  Top-1 Retrieved: {top_1_paper_id}")
    print(f"  P@1: {p1}")
    print(f"  Top-5 Results:")
    for rank, (_, row) in enumerate(results.iterrows(), 1):
        dist = row.get("_distance", "N/A")
        marker = "✅" if row["paper_id"] in ground_truth_list else "❌"
        if isinstance(dist, float):
            print(f"    {rank}. {row['paper_id']} - {row['title'][:60]}... (distance: {dist:.4f}) {marker}")
        else:
            print(f"    {rank}. {row['paper_id']} - {row['title'][:60]}... {marker}")

# Compute and display average P@1
avg_p_at_1 = np.mean(p_at_1_scores)
print("\n" + "=" * 80)
print("BASELINE P@1 RESULTS SUMMARY")
print("=" * 80)

results_summary = pd.DataFrame({
    "user_id": [u["persona_id"] for u in users_data],
    "name": [u["name"] for u in users_data],
    "initial_query": [u["initial_query"] for u in users_data],
    "ground_truth": [str([p["paper_id"] for p in u["relevant_papers"]]) for u in users_data],
    "p_at_1": p_at_1_scores
})
display(results_summary)

print(f"\n📊 Average P@1 (Baseline - Initial Queries): {avg_p_at_1:.4f}")
print(f"   Correct retrievals: {sum(p_at_1_scores)} / {len(p_at_1_scores)}")
````

````python
# WRITE YOUR CODE HERE
# Define user simulator function that responds to clarifying questions
# based ONLY on the user's information_need (not the full persona)

def simulate_user_response(
    llm: ChatOpenAI,
    information_need: str,
    conversation_history: List[Dict[str, str]],
    clarifying_question: str
) -> str:
    """
    Simulate a user (researcher) responding to a clarifying question from the
    search assistant. Uses ONLY the information_need to generate responses.
    
    Args:
        llm: Language model instance
        information_need: The user's detailed information need description
        conversation_history: List of previous conversation turns
        clarifying_question: The current clarifying question from the assistant
    
    Returns:
        Simulated user response string
    """
    # Build conversation history string
    history_str = ""
    for turn in conversation_history:
        if turn["role"] == "assistant":
            history_str += f"Assistant: {turn['content']}\n"
        elif turn["role"] == "user":
            history_str += f"User: {turn['content']}\n"
    
    # Convert information_need to string if it's a dict
    if isinstance(information_need, dict):
        info_need_str = json.dumps(information_need, ensure_ascii=False, indent=2)
    else:
        info_need_str = str(information_need)
    
    # System prompt for user simulation
    system_prompt = """You are simulating a researcher who is looking for a specific scientific paper. 
You have a specific information need in mind. When the search assistant asks you clarifying questions, 
respond naturally and helpfully based on your information need.

Your information need is:
{information_need}

Rules:
- Answer the assistant's questions based ONLY on the information described in your information need
- Be natural and conversational, like a real researcher would respond
- Provide specific details when asked, but only those consistent with your information need
- Do not make up information that is not related to your information need
- Keep responses concise but informative (2-4 sentences)
- Respond in English"""
    
    # Build messages
    messages = [
        SystemMessage(content=system_prompt.format(information_need=info_need_str))
    ]
    
    # Add conversation history
    for turn in conversation_history:
        if turn["role"] == "assistant":
            messages.append(AIMessage(content=turn["content"]))
        elif turn["role"] == "user":
            messages.append(HumanMessage(content=turn["content"]))
    
    # Add the current clarifying question
    messages.append(AIMessage(content=clarifying_question))
    
    # Generate response
    response = llm.invoke(messages)
    return response.content.strip()


# Test the user simulator with a sample
print("--- Testing User Simulator ---")
test_info_need = users_data[0].get("information_need", users_data[0].get("initial_query", ""))
test_question = "Could you tell me more about what specific aspect of this topic you are interested in?"

test_response = simulate_user_response(
    llm=llm,
    information_need=test_info_need,
    conversation_history=[],
    clarifying_question=test_question
)
# Convert info_need to string for display since it may be a dict
test_info_need_str = json.dumps(test_info_need, ensure_ascii=False)[:150] if isinstance(test_info_need, dict) else str(test_info_need)[:150]
print(f"Information Need: {test_info_need_str}...")
print(f"Clarifying Question: {test_question}")
print(f"Simulated Response: {test_response}")
print("\n✅ User simulator function defined successfully!")
````

````python
# WRITE YOUR CODE HERE
# Build conversational search system with clarifying questions

# System prompt for the conversational search assistant
ASSISTANT_SYSTEM_PROMPT = """You are an expert research assistant helping a researcher find a specific scientific paper.
Your goal is to ask clarifying questions to better understand the researcher's information need.

Rules:
- Ask ONE focused clarifying question at a time
- Your questions should help narrow down the specific paper the researcher is looking for
- Ask about: specific methods, datasets, application domains, key contributions, authors, or publication venue
- Be concise and professional
- Ask questions in English
- After gathering enough information (usually 2-3 rounds), indicate you have enough information by starting your message with "[DONE]"
- Do NOT ask more than 3 clarifying questions total"""

MAX_CONVERSATION_ROUNDS = 3  # Maximum number of clarifying question rounds


def run_conversational_search(
    llm: ChatOpenAI,
    user: Dict[str, Any],
    max_rounds: int = MAX_CONVERSATION_ROUNDS
) -> Dict[str, Any]:
    """
    Run a full conversational search session for one user.
    The assistant asks clarifying questions, the simulated user responds,
    and the conversation history is stored.
    
    Args:
        llm: Language model instance
        user: User data dictionary containing initial_query, information_need, etc.
        max_rounds: Maximum number of conversation rounds
    
    Returns:
        Dictionary with user_id, conversation_history, and full_conversation_text
    """
    user_id = user.get("user_id", user.get("id", "unknown"))
    initial_query = user["initial_query"]
    information_need = user.get("information_need", initial_query)
    
    conversation_history = []
    all_clarifying_questions = []
    
    # Start with the initial query from the user
    conversation_history.append({
        "role": "user",
        "content": initial_query
    })
    
    for round_num in range(max_rounds):
        # Build messages for the assistant
        assistant_messages = [
            SystemMessage(content=ASSISTANT_SYSTEM_PROMPT)
        ]
        
        for turn in conversation_history:
            if turn["role"] == "user":
                assistant_messages.append(HumanMessage(content=turn["content"]))
            elif turn["role"] == "assistant":
                assistant_messages.append(AIMessage(content=turn["content"]))
        
        # Generate clarifying question from assistant
        assistant_response = llm.invoke(assistant_messages)
        assistant_text = assistant_response.content.strip()
        
        # Check if assistant thinks it has enough information
        if assistant_text.startswith("[DONE]"):
            conversation_history.append({
                "role": "assistant",
                "content": assistant_text
            })
            break
        
        conversation_history.append({
            "role": "assistant",
            "content": assistant_text
        })
        all_clarifying_questions.append(assistant_text)
        
        # Simulate user response
        user_response = simulate_user_response(
            llm=llm,
            information_need=information_need,
            conversation_history=conversation_history[:-1],  # exclude last assistant msg
            clarifying_question=assistant_text
        )
        
        conversation_history.append({
            "role": "user",
            "content": user_response
        })
    
    # Build full conversation text for search
    full_conversation_text = " ".join([
        turn["content"] for turn in conversation_history
    ])
    
    return {
        "user_id": user_id,
        "initial_query": initial_query,
        "conversation_history": conversation_history,
        "clarifying_questions": all_clarifying_questions,
        "full_conversation_text": full_conversation_text,
        "num_rounds": len(all_clarifying_questions)
    }


# Run conversational search for ALL users
print("=" * 80)
print("RUNNING CONVERSATIONAL SEARCH FOR ALL USERS")
print("=" * 80)

all_conversations = []

for i, user in enumerate(users_data):
    user_id = user.get("user_id", user.get("id", i))
    print(f"\n--- Processing User {user_id} ({i+1}/{len(users_data)}) ---")
    
    conversation_result = run_conversational_search(llm=llm, user=user)
    all_conversations.append(conversation_result)
    
    # Display the conversation
    print(f"  Initial Query: {conversation_result['initial_query'][:80]}...")
    print(f"  Number of rounds: {conversation_result['num_rounds']}")
    for turn in conversation_result["conversation_history"]:
        role_label = "🧑 User" if turn["role"] == "user" else "🤖 Assistant"
        print(f"  {role_label}: {turn['content'][:120]}...")
    
    # Small delay to avoid rate limiting
    time.sleep(1)

print(f"\n✅ Conversational search completed for all {len(all_conversations)} users!")

# Save conversations to CSV
conversations_df = pd.DataFrame([
    {
        "user_id": conv["user_id"],
        "initial_query": conv["initial_query"],
        "num_rounds": conv["num_rounds"],
        "full_conversation": conv["full_conversation_text"][:500],
        "conversation_json": json.dumps(conv["conversation_history"], ensure_ascii=False)
    }
    for conv in all_conversations
])
conversations_df.to_csv("conversations.csv", index=False, encoding="utf-8")
print("Conversations saved to conversations.csv")
````

````python
# WRITE YOUR CODE HERE
# Rewrite the initial query using conversation history with LLM

QUERY_REWRITE_SYSTEM_PROMPT = """You are an expert query rewriting system for academic paper search.
Given an initial search query and a conversation between a researcher and a search assistant, 
rewrite the initial query to be more specific and informative based on the information gathered 
during the conversation.

Rules:
- Incorporate key details from the conversation into the rewritten query
- Make the query specific enough to find the exact paper the researcher needs
- Keep the rewritten query concise but comprehensive (1-3 sentences)
- Focus on: topic, methods, datasets, key contributions, application domain
- Write the rewritten query in English
- Return ONLY the rewritten query, nothing else"""


def rewrite_query(
    llm: ChatOpenAI,
    initial_query: str,
    conversation_history: List[Dict[str, str]]
) -> str:
    """
    Rewrite the initial query using information from the conversation history.
    
    Args:
        llm: Language model instance
        initial_query: The original user query
        conversation_history: List of conversation turns
    
    Returns:
        Rewritten query string
    """
    # Format conversation history
    conv_text = ""
    for turn in conversation_history:
        role = "User" if turn["role"] == "user" else "Assistant"
        conv_text += f"{role}: {turn['content']}\n"
    
    messages = [
        SystemMessage(content=QUERY_REWRITE_SYSTEM_PROMPT),
        HumanMessage(content=f"Initial Query: {initial_query}\n\nConversation:\n{conv_text}\n\nRewritten Query:")
    ]
    
    response = llm.invoke(messages)
    return response.content.strip()


# Rewrite queries for all users
print("=" * 80)
print("QUERY REWRITING USING CONVERSATION HISTORY")
print("=" * 80)

rewritten_queries = []

for conv in all_conversations:
    user_id = conv["user_id"]
    initial_query = conv["initial_query"]
    history = conv["conversation_history"]
    
    # Rewrite the query
    rewritten = rewrite_query(llm=llm, initial_query=initial_query, conversation_history=history)
    rewritten_queries.append({
        "user_id": user_id,
        "initial_query": initial_query,
        "rewritten_query": rewritten
    })
    
    print(f"\nUser {user_id}:")
    print(f"  Initial Query:   {initial_query[:100]}...")
    print(f"  Rewritten Query: {rewritten[:150]}...")
    
    time.sleep(0.5)

# Save rewritten queries
rewritten_df = pd.DataFrame(rewritten_queries)
rewritten_df.to_csv("rewritten_queries.csv", index=False, encoding="utf-8")
print(f"\n✅ Rewritten {len(rewritten_queries)} queries")
print("Saved to rewritten_queries.csv")
display(rewritten_df)
````

````python
# WRITE YOUR CODE HERE
# Evaluate quality of clarifying questions using LLM as judge + human annotation

# Step 1: Collect all clarifying questions from conversations
all_clarifying_qs = []
for conv in all_conversations:
    user_id = conv["user_id"]
    initial_query = conv["initial_query"]
    for i, q in enumerate(conv["clarifying_questions"]):
        all_clarifying_qs.append({
            "user_id": user_id,
            "initial_query": initial_query,
            "question_index": i,
            "clarifying_question": q
        })

print(f"Total clarifying questions generated: {len(all_clarifying_qs)}")

# Step 2: Randomly sample 20% for annotation
sample_size = max(1, int(len(all_clarifying_qs) * 0.2))
random.seed(42)
sampled_indices = random.sample(range(len(all_clarifying_qs)), sample_size)
sampled_questions = [all_clarifying_qs[i] for i in sampled_indices]
print(f"Sampled {sample_size} questions for evaluation (20%)")

# Step 3: Human annotation FIRST - interactive input from user
# You (the human annotator) must rate each sampled question on 4 criteria (1-5)
print("\n" + "=" * 80)
print("STEP 1: HUMAN ANNOTATION - Please rate each question on a scale of 1-5")
print("=" * 80)
print("Criteria:")
print("  1. Naturalness: Is the question natural and grammatically correct? (1-5)")
print("  2. Relevance: Is the question relevant to the user's need? (1-5)")
print("  3. Usefulness: Would the answer help improve search results? (1-5)")
print("  4. Overall Quality: Considering all aspects, overall quality? (1-5)")
print("=" * 80)

human_scores = []
for idx, sq in enumerate(sampled_questions):
    print(f"\n--- Question {idx+1}/{sample_size} ---")
    print(f"User ID: {sq['user_id']}")
    print(f"Initial Query: {sq['initial_query']}")
    print(f"Clarifying Question: {sq['clarifying_question']}")
    print()
    
    while True:
        try:
            # Single combined input to avoid multiple input() calls blocking in Jupyter
            ratings_str = input(f"  Enter 4 ratings (Naturalness Relevance Usefulness Overall) separated by space: ")
            parts = ratings_str.strip().split()
            if len(parts) != 4:
                print("  ⚠️ Please enter exactly 4 numbers separated by spaces. Try again.")
                continue
            nat, rel, use, ovr = int(parts[0]), int(parts[1]), int(parts[2]), int(parts[3])
            if all(1 <= v <= 5 for v in [nat, rel, use, ovr]):
                break
            else:
                print("  ⚠️ All values must be between 1 and 5. Try again.")
        except ValueError:
            print("  ⚠️ Please enter integer values between 1 and 5. Try again.")
    
    human_scores.append({
        "naturalness": nat,
        "relevance": rel,
        "usefulness": use,
        "overall_quality": ovr
    })
    print(f"  ✅ Saved: nat={nat}, rel={rel}, use={use}, ovr={ovr}")

print(f"\n✅ Human annotation completed for {len(human_scores)} questions")

# Step 4: LLM-based evaluation on the SAME sampled questions
print("\n" + "=" * 80)
print("STEP 2: LLM EVALUATION")
print("=" * 80)

LLM_JUDGE_PROMPT = """You are an expert evaluator of clarifying questions in a conversational search system.
Evaluate the following clarifying question on four criteria. Rate each on a scale of 1-5.

Initial Query: {initial_query}
Clarifying Question: {clarifying_question}

Criteria:
1. Naturalness (1-5): Is the question natural, fluent, and grammatically correct?
2. Relevance (1-5): Is the question relevant to the user's information need?
3. Usefulness (1-5): Would the answer help improve search results?
4. Overall Quality (1-5): Considering all aspects, what is the overall quality?

Return ONLY a JSON object like this:
{{"naturalness": X, "relevance": X, "usefulness": X, "overall_quality": X}}"""

llm_scores = []
for sq in sampled_questions:
    messages = [
        HumanMessage(content=LLM_JUDGE_PROMPT.format(
            initial_query=sq["initial_query"],
            clarifying_question=sq["clarifying_question"]
        ))
    ]
    
    response = llm.invoke(messages)
    raw_text = response.content.strip()
    
    # Parse JSON response
    try:
        # Clean markdown fences if present
        cleaned = raw_text.strip()
        if cleaned.startswith("```"):
            import re
            cleaned = re.sub(r'^```(?:json)?\s*', '', cleaned)
            cleaned = re.sub(r'```\s*$', '', cleaned)
        scores = json.loads(cleaned)
    except json.JSONDecodeError:
        # Fallback: default scores
        scores = {"naturalness": 3, "relevance": 3, "usefulness": 3, "overall_quality": 3}
    
    llm_scores.append(scores)
    time.sleep(0.5)

print(f"LLM evaluation completed for {len(llm_scores)} questions")

# Step 5: Compute Fleiss' Kappa for each criterion
criteria = ["naturalness", "relevance", "usefulness", "overall_quality"]
kappa_results = {}

print("\n" + "=" * 80)
print("LLM AS JUDGE - EVALUATION RESULTS")
print("=" * 80)

# Display individual scores
eval_df = pd.DataFrame({
    "question": [sq["clarifying_question"][:80] + "..." for sq in sampled_questions],
    "llm_naturalness": [s["naturalness"] for s in llm_scores],
    "llm_relevance": [s["relevance"] for s in llm_scores],
    "llm_usefulness": [s["usefulness"] for s in llm_scores],
    "llm_overall": [s["overall_quality"] for s in llm_scores],
    "human_naturalness": [s["naturalness"] for s in human_scores],
    "human_relevance": [s["relevance"] for s in human_scores],
    "human_usefulness": [s["usefulness"] for s in human_scores],
    "human_overall": [s["overall_quality"] for s in human_scores],
})
display(eval_df)

# Compute Fleiss' Kappa using aggregate_raters
print("\n--- Fleiss' Kappa Scores ---")
for criterion in criteria:
    llm_ratings = [s[criterion] for s in llm_scores]
    human_ratings = [s[criterion] for s in human_scores]
    
    # Build matrix: each row is an item, each column shows how many raters gave that category
    # We have 2 raters, ratings from 1-5
    rating_matrix = []
    for llm_r, human_r in zip(llm_ratings, human_ratings):
        row = [0] * 5  # categories 1-5
        row[llm_r - 1] += 1
        row[human_r - 1] += 1
        rating_matrix.append(row)
    
    rating_matrix = np.array(rating_matrix)
    
    # Compute Fleiss' Kappa
    try:
        kappa = fleiss_kappa(rating_matrix, method='fleiss')
    except Exception:
        # Fallback to Cohen's Kappa for 2 raters
        kappa = cohen_kappa_score(llm_ratings, human_ratings, weights='linear')
    
    kappa_results[criterion] = kappa

# Display Kappa results table
print("\n" + "=" * 60)
kappa_df = pd.DataFrame({
    "Criterion": ["Naturalness", "Relevance", "Usefulness", "Overall Quality"],
    "Fleiss' Kappa (κ)": [kappa_results[c] for c in criteria],
    "Interpretation": [
        "Poor" if kappa_results[c] < 0.20 else
        "Fair" if kappa_results[c] < 0.40 else
        "Moderate" if kappa_results[c] < 0.60 else
        "Good" if kappa_results[c] < 0.80 else
        "Excellent"
        for c in criteria
    ]
})
display(kappa_df)

# Average LLM scores
print("\n--- Average LLM Scores ---")
avg_llm = {c: np.mean([s[c] for s in llm_scores]) for c in criteria}
for c in criteria:
    print(f"  {c}: {avg_llm[c]:.2f}")

# Average Human scores
print("\n--- Average Human Scores ---")
avg_human = {c: np.mean([s[c] for s in human_scores]) for c in criteria}
for c in criteria:
    print(f"  {c}: {avg_human[c]:.2f}")

# Save evaluation results
eval_results = {
    "sampled_questions": sampled_questions,
    "llm_scores": llm_scores,
    "human_scores": human_scores,
    "kappa_results": {k: float(v) for k, v in kappa_results.items()}
}
with open("evaluation_results.json", "w", encoding="utf-8") as f:
    json.dump(eval_results, f, ensure_ascii=False, indent=2)
print("\n✅ Evaluation results saved to evaluation_results.json")
````

The 50% failure rate highlights the **Vocabulary Gap** and **Polysemy** problems in IR.

<h2 style="color: #24292e; border-bottom: 1px solid #eaecef; padding-bottom: 8px; margin-top: 30px;">
Part 2: Conversational Search System
</h2>

<blockquote style="background-color: #fff8c5; border-left: 5px solid #e3b341; padding: 15px; margin-left: 0; margin-right: 0;">
  <strong>Objective:</strong> To elicit the user's latent information need through multi-turn dialogue and utilize this context to improve retrieval accuracy.
</blockquote>

### 1. System Architecture Implementation
We built a closed-loop simulation:
1.  **User Simulator:** An LLM prompted with a hidden `information_need` profile to answer questions naturally.
2.  **Search Assistant:** An LLM agent prompted to ask up to 3 clarifying questions (e.g., asking about datasets, methods, or venues).
3.  **Conversation Logging:** We captured the full history (User Query $\to$ Clarification $\to$ Response).

### 2. Strategy A: Search with Full Conversation Text
We concatenated the entire dialogue history into a single string and generated an embedding for retrieval.

*   **Average P@1:** **0.8500** (85%)
*   **Improvement:** +0.35 over baseline.
*   **Analysis:**
    *   **Success Factors:** The dialogue introduced critical discriminative terms. For User 14 (Medical Imaging), the conversation added terms like *"U-Net"*, *"segmentation"*, and *"survey"*. This shifted the query embedding vector from a general "medical" cluster to a specific "segmentation architecture" sub-cluster.
    *   **Limitations:** Raw conversation contains **semantic noise** (e.g., *"Hello"*, *"Thank you"*, *"Could you clarify"*). These tokens add dimensions to the vector that are irrelevant to the document content, potentially diluting the similarity score.

### 3. Strategy B: Search with Rewritten Queries (Optimal)
We implemented a **Query Rewriting** module. An LLM processed the conversation history to generate a standalone, semantically dense search query (e.g., *"Survey papers (2020–2024) on deep learning–based medical image segmentation..."*).

*   **Average P@1:** **0.9500** (95%)
*   **Correct Retrievals:** 19 / 20
*   **Improvement:** +0.45 over baseline and +0.10 over raw conversation.

**Why Rewriting Wins:**
Query rewriting acts as both a **Filter** (removing conversational pleasantries) and an **Amplifier** (explicitly stating constraints like dates or venues). It creates a "Canonical Query" that aligns perfectly with the linguistic structure of academic abstracts.

**Analysis of the Single Failure (User 7 - David Johnson):**
*   **Query:** "reinforcement learning"
*   **Target Paper:** *Training language models to follow instructions with human feedback* (RLHF foundation).
*   **Retrieved Paper:** *Deep Reinforcement Learning from Human Preferences*.
*   **Root Cause:** This is a **Semantic Collision**. Both papers discuss "RLHF", "Human Preferences", and "Language Models". They are semantically nearly identical in the embedding space. The model retrieved a highly relevant paper, just not the *exact* ground truth ID. This indicates the system works, but the ground truth definition in this specific instance was extremely narrow.

خوانایی و دقت بررسی‌ها در گزارش نهایی از اهمیت ویژه‌ای برخوردار است. به تمرین‌هایی که به صورت کاغذی تحویل داده شوند یا به صورت عکس در سایت بارگذاری شوند، ترتیب اثری داده نخواهد شد.</li>
<li>
 همه‌ی کدهای پیوست گزارش بایستی قابلیت اجرای مجدد داشته باشند. در صورتی که برای اجرا مجدد آن‌ها نیاز به تنظیمات خاصی می‌باشد، بایستی تنظیمات مورد نیاز را نیز در گزارش خود ذکر کنید.  دقت کنید که  تمامی کدها باید توسط شما اجرا شده باشند و نتایج اجرا در فایل کدهای ارسالی مشخص باشد. به کدهایی که نتایج اجرای آن‌ها در فایل ارسالی مشخص نباشد نمره‌ای تعلق نمی‌گیرد.
</li>
<li>توجه کنید این تمرین باید به صورت تک‌نفره انجام شود و پاسخ‌های ارائه شده باید نتیجه فعالیت فرد نویسنده باشد (همفکری و به اتفاق هم نوشتن تمرین نیز ممنوع است). در صورت مشاهده
 تشابه به همه افراد مشارکت‌کننده، نمره تمرین صفر و به استاد گزارش می‌گردد.
 </li>

 <li>
لطفاً تمامی پاسخ‌های متنی خود را با <b>فونت وزیر (Vazir)</b> و به‌صورت <b>راست‌چین</b> بنویسید.  
از استفاده از فونت‌های پیش‌فرض خودداری کنید تا ظاهر نوت‌بوک شما یک‌دست و خوانا باشد.  
در بخش‌های تشریحی، سعی کنید پاسخ‌ها را کامل، منسجم و با رعایت نگارش فارسی بنویسید.  
همچنین، به چینش تمیز سلول‌ها و اجرای درست کدها توجه کنید تا تمرین شما با فرمت خواسته‌شده و استاندارد ارائه شود.
</li>
